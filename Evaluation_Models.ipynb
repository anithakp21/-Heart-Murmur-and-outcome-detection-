# Defining a function to evaluate

```python
def takeinput(csvfile,folder):
    df_model = pd.read_csv(csvfile)
    df = df_model.copy()
    df_id = df['Id']
    df_mur = df[['Present','Unknown','Absent']]
    df_ob = df[['Abnormal','Normal']]

#Creating new dataframe for murmurs
    df_new_mur = pd.DataFrame(columns=['Present','Unknown', 'Absent'])
    for i in range(len(df_mur)):
      temp = df_mur.iloc[i].tolist()
      labels = [0,0,0]
      labels[np.argmax(df_mur.loc[i].tolist())] = 1
      labels = [int(x) for x in labels]
      df_new_mur.loc[len(df_new_mur)] = labels
      df_new_mur.loc[len(df_new_mur)] = temp

#Creating a new dataframe for outcomes
    df_new_ob = pd.DataFrame(columns=['Abnormal', 'Normal'])
    for i in range(len(df_ob)):
      temp = df_ob.iloc[i].tolist()
      labels = [0,0]
      labels[np.argmax(df_ob.loc[i].tolist())] = 1
      labels = [int(x) for x in labels]
      df_new_ob.loc[len(df_new_ob)] = labels
      df_new_ob.loc[len(df_new_ob)] = temp

    df_new_id = df_id.repeat(2).reset_index(drop=True)
    df_new_id = df_id.repeat(2).astype(int).reset_index(drop=True) ## Convert IDs to int
    df_major = pd.concat([df_new_id, df_new_mur, df_new_ob], axis=1)
    df_csv = df_major[['Present','Unknown','Absent','Abnormal','Normal']]

    label1 = df_csv.iloc[0].tolist()
    label2 = df_csv.iloc[0+1].tolist()
    data = [label1,label2]
    new_data = [['#' + str(df_new_id[i])]]
    new_data.extend(data)
    data = new_data

    counter = 0
    for i in range(0,len(df_csv),2):
      counter+=1
      label1 = df_csv.iloc[i].tolist()
      label3 = ['Present','Unknown','Absent','Abnormal','Normal']
      label2 = df_csv.iloc[i+1].tolist()
      data = [label3,label1,label2]
      new_data = [['#' + str(df_new_id[i])]]
      new_data.extend(data)
      data = new_data
      file = open('{}/{}.csv'.format(folder, df_new_id[i]), 'w', newline='')
      writer = csv.writer(file)
      for row in data:
          writer.writerow(row)
      file.close()
      
    print(counter)
 
    
```
